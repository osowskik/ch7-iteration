```python
#For some reason this works nearly OK when I run it through jupyter notebook, but not when I run it through terminal. 
#When running through jupyter notebook it gives me an error about the turtle's name, 
# but when I run it again and answer the input questions again the same way, it's OK
# A module for drawing a chart with the turtle
import turtle  # import module

File = input("Enter your filename: ")
Title = input ("Enter your title: ")

# Define window size as constants
window = turtle.Screen()  # create a window for the turtle to draw on
window.title(Title)  # the title to show at the top of the window
WINDOW_WIDTH = 1000  # size constants for easy changing
WINDOW_HEIGHT = 700
window.setup(WINDOW_WIDTH, WINDOW_HEIGHT)  # specify window size (width, height)
window.setworldcoordinates(0, 0, WINDOW_WIDTH, WINDOW_HEIGHT)  # coordinate system: origin at lower-left

# Create the turtle
shelly = turtle.Turtle() #for some reason this gives an error and then when rerun it's ok
shelly.shape("turtle")
shelly.speed("fast")  # how fast the turtle should move

#Need to loop through the file in order to count the number of observations
#observations take 3 lines
#should be 14 observations
def count_observations(File):
    with open(File, 'r') as my_file:
        for i, l in enumerate(my_file):
                pass
    return (i + 1)/3
count_observations(r"/Users/kayleeosowski/Google_Drive/LIS_511/Assignment3/a3-turtle-charter-osowskik/data/huskies2016.txt")

def get_max_value(File):
# read through the file one observation (3 lines) at a time
    values = []
    file = open(File, 'r')
    lines = file.readlines()
    for i, line in enumerate(lines):
        #if index is divisible by three, it's a match
        if (i + 1) % 3 == 0:
            name = lines[i - 2].replace('\n', '')
            feature_1 = lines[i - 1].replace('\n', '')
            feature_2 = line.replace('\n', '')
            values.append([feature_1, feature_1])  
# # #keep track of the "biggest" value you've found so far
# # # #At each observation, check whether the feature of interest is greater than the current biggest, 
# # # #and if so make it become the new biggest!          
    king = values[0]
    for value in values:
        if value[1] > king[1]:
            king = value
    return king[0]
get_max_value(r"/Users/kayleeosowski/Google_Drive/LIS_511/Assignment3/a3-turtle-charter-osowskik/data/huskies2016.txt")

#def draw_axis():  
    #x axis
shelly.penup()
shelly.right(270)
shelly.forward(40)
shelly.right(90)
shelly.forward(40)
shelly.pendown() 
shelly.forward(950)
shelly.right(180) 
shelly.forward(950)
#y axis 
shelly.right(90)
shelly.forward(600)
shelly.right(180)
#draw_axis()

#def draw_y_tick_mark():
#y axis is 600, need 10 tickmarks, need to move turtle down by 60 10 times
#first tickmark is at the top
shelly.right(90)
shelly.forward(10)
shelly.right(180)
shelly.forward(10)
shelly.right(90)

#next 9 tickmarks need to repeat
for x in range(9):
    shelly.forward(60)
    shelly.right(90)
    shelly.forward(10)
    shelly.right(180)
    shelly.forward(10)
    shelly.right(90)
   
   #setting font
FONTSIZE = 12
FONT = ('Arial', FONTSIZE, 'normal')
shelly.penup()

#writing number labels
shelly.goto(10,90)
shelly.right(180)
shelly.write('7', font=FONT)

y_labels = [14, 21, 28, 35, 42, 49, 56, 63, 70]
for i in range(9):
    shelly.forward(60)
    shelly.write(y_labels[i], font=FONT)

shelly.goto(0,20)
shelly.right(90)

# x_labels
# total_bar_space = 10 * 14 
# total_bar_width = 900 - total_bar_space
# width = total_bar_width/14
with open(File, 'r') as f:
    for line in f.read().split("\n")[0::3]:
        shelly.forward(67)
        shelly.write(line, font=FONT)

shelly.goto(50, 40)
shelly.left(90)

import random
colors = ["#804575", "#ffd6f8", "#ff8ae9", "#806b7c", "#cc6ebb", "#6c4780", "#f2d9ff", "#d78cff", "#796c80", "#ac70cc", "#803360", "#ffb3df", "#805970", "#cc5299"]
def draw_bar(turtle, File):
    with open(File, 'r') as f:
        for line in f:
            scores = f.read().split("\n")[0::3]
            height = [8.57 * int(i) for i in scores]
            rounded_height = [round(x,2) for x in height]
        for i in rounded_height:
                turtle.fillcolor(random.choice(colors))
                shelly.begin_fill()
                shelly.pendown()
                shelly.forward(i)
                shelly.right(90)
                shelly.forward(57)
                shelly.right(90)
                shelly.forward(i)
                shelly.penup()
                shelly.left(90)
                shelly.forward(10)
                shelly.left(90)
                shelly.pendown()
                shelly.end_fill()
draw_bar(shelly, r"/Users/kayleeosowski/Google_Drive/LIS_511/Assignment3/a3-turtle-charter-osowskik/data/huskies2016.txt")
              
# Make the turtle graphics appear and run, waiting for the user to close the screen
# This MUST be the last statement executed in the script
window.mainloop()
```
